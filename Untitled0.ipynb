```python
#modul Logistic Regression
from sklearn.linear_model import LogisticRegression

#modul Naive Bayes
from sklearn.naive_bayes import GaussianNB


#modul encoder -> variabel kategorikal
from sklearn.preprocessing import OrdinalEncoder

#modul split data
from sklearn.model_selection import train_test_split

#modul metric (Akurasi)
from sklearn.metrics import accuracy_score, recall_score, precision_score

#modul untuk import dataset
import pandas as pd
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Import Dataset

```python
df=pd.read_csv('bank.csv')
df['loan']=np.where(df['loan']=='yes',1,0)
df
```

# Define Feature & Label

```python
#featue
X=df[['age', 'job', 'marital', 'education', 'loan', 'duration']]
#label
y=df['deposit']
```

# Endcode

```python
encoder=OrdinalEncoder()
for i in X.select_dtypes(exclude='number').columns:
    X[i]=encoder.fit_transform(X[[i]])
X
```

# Split Data

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=10)

#train -> untuk model latihan, test -> untuk evaluasi model
```

# Modeling

# Logistic Regresion

```python
#create model
logreg=LogisticRegression()
logreg = LogisticRegression(class_weight='balanced')

# fit model (model latihan pada data train)
logreg.fit(X_train,y_train)
```

```python
#prediksi---
y_prob_log_train = logreg.predict_proba(X_train)[:, 1]
y_prob_log_test = logreg.predict_proba(X_test)[:, 1]
y_pred_logreg_train = np.where(y_prob_log_train > 0.35, 'yes', 'no')
y_pred_logreg_test = np.where(y_prob_log_test > 0.35, 'yes', 'no')
```

# Native Bayes

```python
#create model
nb=GaussianNB()

# fit model (model latihan pada data train)
nb.fit(X_train,y_train)
```

```python
# prediksi --
y_prob_nb_train = nb.predict_proba(X_train)[:, 1]
y_prob_nb_test = nb.predict_proba(X_test)[:, 1]

y_pred_nb_train = np.where(y_prob_nb_train > 0.3, 'yes', 'no')
y_pred_nb_test = np.where(y_prob_nb_test > 0.3, 'yes', 'no')
```

# Cek Akurasi Setiap Model

```python
list_y_train=[y_pred_logreg_train,y_pred_nb_train]
nama_model=['Logistic Regression','Naive Bayes']
list_kosong_train=[]
for i in list_y_train:
    accuracy=str(round(100*accuracy_score(y_train,i),2))+' %'
    list_kosong_train.append(accuracy)

list_kosong_test=[]
list_y_test=[y_pred_logreg_test,y_pred_nb_test]
for i in list_y_test:
    accuracy=str(round(100*accuracy_score(y_test,i),2))+' %'
    list_kosong_test.append(accuracy)

pd.DataFrame({'Nama Model':nama_model,'Akurasi Train':list_kosong_train,'Akurasi Test':list_kosong_test})
```

# Check Recall Setiap Model

```python
# 1. Fungsi pembantu buat pastiin semuanya jadi angka
def force_int(data):
    return pd.Series(data).replace({'yes': 1, 'no': 0}).astype(int).values

nama_model = ['Logistic Regression', 'Naive Bayes']

# EVALUASI TRAIN
list_y_train = [y_pred_logreg_train, y_pred_nb_train]
list_kosong_train = []
for i in list_y_train:
    # Paksa y_train DAN hasil prediksi (i) jadi angka
    recall = str(round(100 * recall_score(force_int(y_train), force_int(i)), 2)) + ' %'
    list_kosong_train.append(recall)

# EVALUASI TEST
list_y_test = [y_pred_logreg_test, y_pred_nb_test]
list_kosong_test = []
for i in list_y_test:
    # Paksa y_test DAN hasil prediksi (i) jadi angka
    recall = str(round(100 * recall_score(force_int(y_test), force_int(i)), 2)) + ' %'
    list_kosong_test.append(recall)

# TAMPILKAN TABEL
pd.DataFrame({'Nama Model': nama_model, 'Recall Train': list_kosong_train, 'Recall Test': list_kosong_test})
```

# Check Prediction Setiap Model


```python
y_train_int = pd.Series(y_train).replace({'yes': 1, 'no': 0}).astype(int)
y_test_int = pd.Series(y_test).replace({'yes': 1, 'no': 0}).astype(int)

def to_int(data):
    return pd.Series(data).replace({'yes': 1, 'no': 0}).astype(int).values

list_y_train = [y_pred_logreg_train, y_pred_nb_train]
nama_model = ['Logistic Regression', 'Naive Bayes']
list_kosong_train = []

for i in list_y_train:
    precision = str(round(100 * precision_score(y_train_int, to_int(i)), 2)) + ' %'
    list_kosong_train.append(precision)

list_kosong_test = []
list_y_test = [y_pred_logreg_test, y_pred_nb_test]

for i in list_y_test:
    precision = str(round(100 * precision_score(y_test_int, to_int(i)), 2)) + ' %'
    list_kosong_test.append(precision)

pd.DataFrame({'Nama Model': nama_model, 'Precision Train': list_kosong_train, 'Precision Test': list_kosong_test})
```
